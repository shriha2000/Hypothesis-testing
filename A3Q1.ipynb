```python
plt.subplots(figsize=(9,6))
plt.subplot(121)
plt.boxplot(cutlets['Unit A'])
plt.title('Unit A')
plt.subplot(122)
plt.boxplot(cutlets['Unit B'])
plt.title('Unit B')
plt.show()
```

```python
plt.subplots(figsize=(9,6))
plt.subplot(121)
plt.hist(cutlets['Unit A'],bins=15)
plt.title('Unit A')
plt.subplot(122)
plt.hist(cutlets['Unit B'],bins=15)
plt.title('Unit B')
plt.show()
```

```python
plt.figure(figsize=(8,6))
label=['Unit A','Unit B']
sns.distplot(cutlets['Unit A'])
sns.distplot(cutlets['Unit B'],hist=True)
plt.legend(labels)
```

```python
sm.qqplot(cutlets["Unit A"],line='q')
plt.title('Unit A')
```

```python
sm.qqplot(cutlets["Unit B"],line='q')
plt.title('Unit B')
```

```python
p_value=stats.ttest_ind(cutlets['Unit A'],cutlets['Unit B'], alternative = 'two-sided')
print('p_values=',p_value)
```
